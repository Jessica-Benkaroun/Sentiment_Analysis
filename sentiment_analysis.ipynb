# Statistical analysis:

Chi-square test: We can use a chi-square test to examine the association between categorical variables in the dataset. For example,we could investigate if there is a significant association between sentiment (positive, negative, neutral) and emotions (joy, sadness, anger, etc.) in the tweets.

We can apply the central limit theorem to our dataset since it is a large dataset which assume our data follow a normal distribution.

To perform the Chi-Square test read and follow the documentation from SciPy:https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html

Follow the exemple given in the documentation, you would need to create a contengency table first of your data of interest before employing the stat test.

A contingency table, also known as a cross-tabulation table or a contingency table of frequencies, is a table that displays the joint distribution of two or more categorical variables. It shows the counts or frequencies of observations falling into various combinations of categories for the variables being studied.

Don't forget to formulate your hypothesis first.

```python
df2= df.copy(deep=True)
```

```python
df2= df.dropna()
```

```python
df2.head()
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table from the two categorical variables
contingency_table = pd.crosstab(df2['sentiment'], df2['emotion'])
contingency_table
```

```python
# Perform the Chi-square test

res = chi2_contingency(contingency_table , lambda_=None)
# Print the test statistics and p-value
print("Chi-square statistic:", res.statistic, "p-value:",  res.pvalue)
```

```python
#  Perform the G-test

res = chi2_contingency(contingency_table , lambda_="log-likelihood")

# Print the test statistic and p-value
print("G-test statistic:", res.statistic, "p-value:", res.pvalue)
```

p-value of zero: when the sample size is small, the statistical power to detect deviations from the null hypothesis is limited. It may require a relatively large deviation (difference between observed data and the expected values under the null hypothesis) from the null hypothesis to obtain a significant result(a small p-value).

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

# Select the relevant columns
df = df[['Text', 'emotion']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Text'], df['emotion'], test_size=0.2, random_state=42)

# Create TF-IDF vectors (Term Frequency-Inverse Document Frequency)
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)

# Train the LinearSVC classifier (ML model)
classifier = LinearSVC() # The LinearSVC classifier is a variant of Support Vector Machines (SVM).
classifier.fit(X_train_vectors, y_train)

# Predict the emotions for the test set
y_pred = classifier.predict(X_test_vectors)

# Evaluate the model
print(classification_report(y_test, y_pred))
```


The classification report helps in understanding the strengths and weaknesses of the classification model and can guide further improvements or adjustments to the model if needed.

precision, recall, and F1-score are reported for each emotion class. The accuracy of the model on the entire test set is also provided. The macro average and weighted average are calculated across all classes to provide an overall performance measure.


Precision is a metric used in classification tasks that measures the accuracy of the positive predictions made by a model. It quantifies the proportion of true positive predictions, higher is better.
Recall is a metric used in classification tasks to measure the ability of a model to correctly identify all positive instances from the actual positive instances.

The F1-score is a metric used in classification tasks that combines precision and recall into a single value, providing a balanced measure of a model's performance.

"Support" refers to the number of instances (samples) in each class present in the testing set or dataset being evaluated. It represents the frequency or count of instances belonging to a specific class.
